````python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculation of mean and median of "Rooms"
mean_rooms = df["Rooms"].mean()
median_rooms = df["Rooms"].median()

# Calculation of skewness
skew_rooms = df["Rooms"].skew()

# Plot the distribution of "Rooms"
plt.figure(figsize=(15,5))
sns.histplot(df["Rooms"], kde = True, bins = 30)
plt.axvline(mean_rooms, color = 'b', linestyle = '--', label = "mean_rooms")
plt.axvline(median_rooms, color = 'r', linestyle = '-', label = "median_rooms")
plt.title("Distribution of data in Rooms")
plt.xlabel("Number of rooms per unit")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Get value of mean_rooms, median_rooms, skewness
mean_rooms, median_rooms, skew_rooms
````

**One-hot encoding**: each category is represented by a separate binary feature.

**Binary encoding**: each category is represented as binary digits (0s and 1s).

**Ordinal encoding**: each unique category is assigned an integer value based on the order or ranking of the categories.

Crime_Rate, Residential_Area, and Highway_Access are represented with categories that are ranked, in ascending order, from "Low" to "High" or "Very Low" to "Very High". Therefore, we will perform ordinal encoding for these ordinal variables.

Next, we need to decide whether to perform one-hot encoding or binary encoding on the nominal variable River. We observed that there are only two categories representing River, "Near" and "Far". In this scenario, binary encoding is appropriate as each category can easily be represented as a binary value of either 0 or 1.
